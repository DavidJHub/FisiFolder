```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import stats
import time
import progressbar
from matplotlib import cm
```

```python
def brillouin(eta,J):
    #m=np.linspace(-J,J,2*J+1)
    #print(m)
    rt=(1/J)*((J+0.5)*(np.cosh(J+0.5)*eta)/np.sin((J+0.5)*eta)-0.5*(np.cosh(0.5*eta)/np.sinh(0.5*eta)))
    return rt
```

```python
brillouin(100,0.5)
```

```python
rannge=np.linspace(-6,6,1000)
plt.plot(rannge,brillouin(rannge,1))
plt.plot(rannge,brillouin(rannge,-1))
```

```python
jey=np.linspace(-1,1,100)
plt.figure(figsize=(12,8))
for i in range(len(jey)):
    plt.plot(rannge,brillouin(rannge,jey[i]))
plt.ylim(-10000,10000)
```

```python
def realbrillouin(eta,J):
    #m=np.linspace(-J,J,2*J+1)
    #print(m)
    rt=((2*J+1)/2*J)*((np.cosh((2*J+1)/2*J)/np.sin((2*J+1)/2*J))-(1/2*J)*(np.cosh((1/2*J)*eta)/np.sinh((1/2*J)*eta)))
    return rt
```

```python
nr=np.linspace(0.00001,6,5000)
jey=np.linspace(0.1,1,7)
plt.figure(figsize=(12,8))
for i in range(len(jey)):
    labels="J="+str(jey[i])
    plt.plot(nr,realbrillouin(nr,jey[i]),label=labels)
plt.ylim(-1,2)
plt.xlim(0,6)
plt.axhline(y=1,color="black",linewidth=2)
plt.legend(loc=8)
```
