```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import scipy.signal
import seaborn
seaborn.set(style='ticks')
```

```python
R=np.linspace(-10,10,100000)

def fun(d,x):
    F=1/(x*np.sqrt(2*d)*1j)*(np.exp(2*x*1j)-np.exp(-2*x*1j))
    return F
```

```python
D=fun(2,R)
```

```python
print(D)
```

```python
delta=np.linspace(0.1,30,30)
plt.figure(figsize=(40,12))
plt.xlim(-10,10)
for k in range (len(delta)):
    string="$\Delta$= " + str(delta[k])
    plt.plot(R,fun(delta[k],R,),label=string,color=cm.twilight_shifted(1/(delta[k])))
    plt.legend(loc=2, prop={'size': 14})
```

```python
plt.figure(figsize=(40,12))
plt.xlim(-10,10)
for k in range (len(delta)):
    string="$\Delta$= " + str(delta[k])
    plt.plot(R,fun(delta[k],R,)**2,label=string,color=cm.twilight_shifted(1/(delta[k])))
    plt.legend(loc=2, prop={'size': 14})
```
