```python
def primes(max):
    ret=[]
    n = 2
    while n<max:
        ret.append(float(n))
        n = gmpy2.next_prime(n)
    return ret

print(primes(30))
```

```python
def PDF(x):
    ret=[]
    for i in range(len(x)):
        if (x[i]<0):
            ret.append((x[i]**(3)))
        else:
            ret.append((x[i]**(-1)))
    return np.asarray(ret)
```

```python
RANGE=np.linspace(-0.1,0.1,100000)
print(PDF(RANGE))
plt.figure(figsize=(60,40))

plt.plot(RANGE,PDF(RANGE),color="red", linewidth=10)
plt.xlim(-0.01,0.01)
plt.ylim(0,1000000000)
plt.grid()
```

```python
def largest_prime_factor(n):
    i = 2
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
    return n
```

```python
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
```
