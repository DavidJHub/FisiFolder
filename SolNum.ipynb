```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def euler(a,aa,aaa,eps):
    
    min_y = 0
    max_y = 10
    N=1000
    h=(max_y-min_y)/N
    y = np.zeros(N)
    y_1 = np.zeros(N)
    y_2 = np.zeros(N)
    y_3 = np.zeros(N)
    y[0]   = a
    y_1[0] = aa
    y_2[0] = aaa
    y_3[0] = 0

    for i in range(1,N):
        
        y_2[i]=y_3[i-1]*h+y_2[i-1]
        
        y[i]=y[i-1]+h
        
        y_1[i]=y_2[i-1]*h+y_1[i-1]
        
        y_3[i]=(-y[i]**2+eps)*y_1[i]
        
    return y,y_1
```

```python
plt.plot(euler(0,0,1,0)[0],euler(0,0,1,0)[1], )
plt.plot(euler(0,1,0,0)[0],euler(0,1,0,0)[1], )
#plt.plot(RKsol(0,1,0)[0],np.cos(2.0*x))
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
```

```python
plt.figure(figsize=(60,40))

for i in range(1,6):
    lab2=str(i)
    plt.plot(euler(0,0,1,i)[0],euler(0,0,1,i)[1],label= "$\epsilon$" + lab)
plt.title("Simétricas",fontsize=80)
plt.legend(fontsize=40)
#plt.xlim(0,10)
    
    
```
